```python
import pandas as pd
import numpy as np
from copy import deepcopy
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn import metrics
df=pd.read_csv("/home/root1/Downloads/zoo_data.csv", header=None)
df = df.iloc[1: , :]
df
```

```python
df.head()
```

```python
X=df.values[:,:16]
Y=df.values[:,-1]
```

```python
X
```

```python
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
clf1 = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Train Decision Tree Classifer
clf1 = clf1.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf1.predict(X_test)
```

```python
print("Accuracy for entropy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix

from sklearn.metrics import accuracy_score
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
Ypredict=model.predict(X_test)
print(accuracy_score(y_test,Ypredict))
print(confusion_matrix(y_test,Ypredict))
```
